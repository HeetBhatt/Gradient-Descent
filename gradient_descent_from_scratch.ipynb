```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
X.shape,y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

using OLS

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='green')
```

Using Gradient Descent

```python
# let's m = 78.35 and b=0
y_pred = ((78.35*X)+0).reshape(4)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),label="OLS",color="green")
plt.plot(X,y_pred,label="b = 0",color="red")
plt.legend()
plt.show()
```

```python
b=-100
m=78.35
epoch =1000
learning_rate = 0.01
for i in range(epoch):
    loss= -2*np.sum(y-m*X.ravel()-b)
    b= b - learning_rate*loss
    y_pred = m*X +b
    plt.plot(X,y_pred)
plt.scatter(X,y) 
```
